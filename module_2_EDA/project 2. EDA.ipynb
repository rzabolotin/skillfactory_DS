## Функции используемые для анализа

```python
def column_summary(column_name):
    """
    Выводит сводюную информацию по колнке
    1) value_counts
    2) количество пустых значений
    3) основные квантили, межквартильный размах и границы выбросов
    """
    print(f'Статистика по полонке {column_name}')
    display(df[column_name].value_counts())
    num_na = sum(df[column_name].isna())
    print(f'Пустых значений {num_na}')

    median = df[column_name].median()
    perc25 = df[column_name].quantile(0.25)
    perc75 = df[column_name].quantile(0.75)
    IQR = perc75 - perc25
    print('25-й перцентиль: {},'.format(perc25),
          '75-й перцентиль: {},'.format(perc75),
          "IQR: {}, ".format(IQR),
          "Границы выбросов: [{f}, {l}].".format(f=perc25 - 1.5*IQR,
                                                 l=perc75 + 1.5*IQR
                                                 ),
          sep='\n'
          )


def display_boxplots(column_name, order=None):
    """
    Выводит бокплоты по колонке
    """
    fig, ax = plt.subplots(figsize=(6, 4))
    sns.boxplot(x=column_name, y='score',
                data=df,
                order=order,
                ax=ax)
    ax.set_title('Boxplot для ' + column_name)
    plt.show()


def find_stat_dif(column_name):
    """
    Ищет статистически значимые отличия в этой колонке
    Если находит, то выводит сообщение на экран
    """
    uniq_values = df[column_name].unique()
    combinations_all = list(combinations(uniq_values, 2))

    for comb in combinations_all:
        A = df[df[column_name] == comb[0]]['score']
        B = df[df[column_name] == comb[1]]['score']

        pvalue = ttest_ind(A, B).pvalue

        if pvalue <= 0.05/len(combinations_all):  # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column_name)
            break
```
